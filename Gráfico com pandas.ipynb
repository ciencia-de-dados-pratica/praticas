# Gráficos usando a biblioteca pandas

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

# Dataframe

```python
velocidade = [0.1, 17.5, 40, 48, 52, 69, 88]
anos_de_vida = [2, 8, 70, 1.5, 25, 12, 28]
animais = ['caracol', 'porco', 'elefante','coelho', 'girafa', 'coiote', 'cavalo']

df = pd.DataFrame({'velocidade': velocidade, 'anos de vida': anos_de_vida}, index=animais)
df
```

# Gráficos

### df.plot.scatter ( )

Cria um gráfico de dispersão.

```python
#sendo c os valores para colorir os pontos
df.plot.scatter(x='velocidade',
                y='anos de vida',);
```

### df.hist ( )

Mostra a distribuição dos dados

```python
#alpha muda a intensidade das cores
#color escolhe a cor
plt.figure();
df.plot.hist(x='anos de vida',alpha=0.5, color = 'g');
```

```python
plt.figure();
df.plot.hist(x='velocidade',alpha=0.5);
```

### df.plot ( ) 

Usando assim ele gera um gráfico de linhas.

```python
df.plot(title = 'Gráfico 1');
```

Subplot divide as informações do gráfico em vários gráficos

```python
df.plot(subplots = True);
```

### plot.barh( )

#### Gráficos com apenas uma característica

```python
df.plot.barh(x='velocidade', title = 'Gráfico de velocidade');
```

```python
df.plot.barh(x='anos de vida', title = 'Gráfico sobre os anos de vida');
```

#### Dividindo as informações em vários gráficos

```python
df.plot.barh(rot = 0, subplots = True);
```

### Juntando as informações em um único gráfico

```python
df.plot.barh();
```

```python
#stacked é para empilhar as barras
df.plot.barh(rot = 0, stacked = True);
```

### plot.bar( )

```python
df.plot.bar(rot = 0, subplots = True);
```

```python
df.plot.bar();
```

```python
df.plot.bar(rot = 0, stacked = True);
```

### plot.pie ( )

Cria um gráfico de pizza.

```python
#figsize determina o tamnho da imagem
df.plot.pie(y='velocidade', figsize=(5, 5));
```

```python
plot = df.plot.pie(subplots=True, figsize=(16, 8))
```
