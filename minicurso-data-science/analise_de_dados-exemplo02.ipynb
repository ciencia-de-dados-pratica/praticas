### Obtenção de Dados


Baixe o conjunto de dados [Pima Indians Diabetes](https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data) do repositório de aprendizagem de máquina UCI.

Mais informações: https://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes

#### Atributos

1. Number of times pregnant 
2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test 
3. Diastolic blood pressure (mm Hg) 
4. Triceps skin fold thickness (mm) 
5. 2-Hour serum insulin (mu U/ml) 
6. Body mass index (weight in kg/(height in m)^2) 
7. Diabetes pedigree function 
8. Age (years) 
9. Class variable (0 or 1)

### Manipulação e Limpeza de Dados

*Data munging* ou *data wrangling* é o processo de converter ou mapear dados de um formato bruto em um outro formato que permita o consumo mais conveniente dos dados com a ajuda de ferramentas semi-automatizadas.

Em nosso caso, vamos considerar que os dados já estão prontos para uso.
